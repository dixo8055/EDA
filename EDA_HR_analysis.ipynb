```python
from google.colab import files
uploaded = files.upload("hranalysis.csv")
```

```python
import numpy as np
import pandas as pd
pd.set_option("display.max_columns", None)
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
HRDataset_v14 = list(uploaded.keys())[0]
df=pd.read_csv(HRDataset_v14)
```

```python
df.info()
```

```python
df['Salary'].sort_values(ascending=False)
```

```python
df_sorted = df.sort_values(by='Salary', ascending=False).head(20)
plt.figure(figsize=(20, 5))
sns.barplot(x=df_sorted.index, y=df_sorted['Salary'], palette='viridis')
plt.xticks(rotation=90)
plt.xlabel("Index")
plt.ylabel("Salary")
plt.title("Salaries in Descending Order")
plt.show()
```

```python
df_sorted = df.sort_values(by='Salary', ascending=False).head(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=df_sorted["Position"], y=df_sorted["Salary"], palette='viridis')
plt.xticks(rotation=90)
plt.xlabel("Position")
plt.ylabel("Salary")
plt.title("Top 20 Salaries by Position")

plt.show()
```
